```python
def get_edge_list(G):
  return [
    {
      'source': G.nodes[edge[0]],
      'target': G.nodes[edge[1]],
      'edge_attributes': G.get_edge_data(edge[0], edge[1])
    }
    for edge in G.edges()
  ]

edge_list = get_edge_list(G)

test_edge = edge_list[0]
print(test_edge['source']['pos'][0])
```
